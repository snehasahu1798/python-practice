```python
# Positional arguments
def sub(a,b):
    print(a - b)
sub(20,10)
```

```python
# keyword arguments
def sum(a,b):
    print(a + b)
sum(a = 10, b = 20)
```

```python
# positional and keyword argument both
def demo(name,msg):
    print("Hello", name,msg)
demo("Sneha", msg = "Good morning")
```

```python
# variable number of positional arguments
def positional(*args):
    print(args)
positional(1,2,3)
```

```python
# variable number of keyword arguments
def keyword(**kwargs):
    print(kwargs)
keyword(name="sneha", age = 20)
```

```python
# global variable
a = 10
def f1():
    print(a)
def f2():
    print(a)
f1()
f2()
```

```python
# local variable
def f11():
    b = 20
    print(b)
    
def f12():
    print(b)
    
f11()
f12()
```
