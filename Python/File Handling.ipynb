<h3>With Statement</h3>

```python
with open("demoFile", 'w') as f:
    f.write("Hi")

with open("demoFile",'r') as f:
    for line in f:
        print(line)
```

```python
data="All Students are STUPIDS"
f=open("abc.txt","w")
f.write(data)
with open("abc.txt","r+") as f:
    text=f.read()
    print(text)
    print("The Current Cursor Position: ",f.tell())
    f.seek(17)
    print("The Current Cursor Position: ",f.tell())
    f.write("GEMS!!!")
    f.seek(0)
    text=f.read()
    print("Data After Modification:")
    print(text)
```

<h3>Writing data to csv file</h3>


```python
import csv
```

```python
with open("emp.csv",'w',newline="") as f:
    
        w = csv.writer(f)
    
        # Header
        w.writerow(["Ename","Emob-number","Eaddress"])
    
        n = int(input("Enter the number of Employees"))
    
        for lines in range(n):
            ename = input("Enter employee name: ")
            emob_number = input("Enter employee mobile number: ")
            eaddress = input("Enter employee address: ")
            
            w.writerow([ename,emob_number,eaddress])
            
print("The employee data is stored inside the csv file")
```

<h3>Reading data from csv file<h3>

```python
f = open("emp.csv",'r')
r = csv.reader(f)
for line in r:
    print(line)
```

<h3>Example - 2</h3>

```python
with open("company.csv",'w',newline="") as f:
    
    w = csv.writer(f)
    w.writerow(["Company_Name","City"])
    
    # Header
    n = int(input("Enter the List of Company"))

    for i in range(n):
        company_name = input("Enter the company name")
        city = input("Enter the city name")

        w.writerow([company_name,city])
        
print("The company data is stored inside the csv file")
    
```

```python
f = open("company.csv",'r')
res = csv.reader(f)
for line in res:
    print(line)
```

<h2>Zipping and Unzipping Files</h2>

<h3>Create a Zip file</h3>

```python
from zipfile import *
f = ZipFile("files.zip",'w',ZIP_DEFLATED)
f.write("abc.txt")
f.write("demo.txt")
f.close()
print("Zip file created successfully !!")
```

<h3>Create a UnZip file</h3>

```python
f = ZipFile("files.zip",'r',ZIP_STORED)
names = f.namelist()
print(names)
```
