<h3>Properties of File Object</h3>

```python
f = open("demoFile",'w')
```

```python
print("File Name: ",f.name)
```

```python
print("File Mode: ",f.mode)
```

```python
print("Is File Readable: ",f.readable())
```

```python
print("Is File Writable: ",f.writable())
```

```python
print("Is File closed: ",f.closed)
```

```python
f.close()
```

```python
print("Is File closed: ",f.closed)
```
