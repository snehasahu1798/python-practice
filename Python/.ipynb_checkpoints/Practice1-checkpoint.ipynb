```python
# To display the sum of first n numbers
n = int(input("Enter the numbers: "))
i = 1
sum = 0
while i <= n:
    sum = sum + i
    i = i + 1
print("The sum of first n numbers is: ", sum)
```

```python
# Write a program to sort the characters of the string and first alphabet symbols
# followed by numeric values

s=input("Enter Some String:")
s1=s2=output=''
for x in s:
    if x.isalpha():
        s1=s1+x
    else:
        s2=s2+x
for x in sorted(s1):
    output=output+x
for x in sorted(s2):
    output=output+x
print(output)
```

```python
# Write a program to find the number of occurrences of each character present in the
# given String

# input: ABCABCABBCDE
# output: A-3,B-4,C-3,D-1,E-1

s = input("Enter the string: ")
d = {}
for i in s:
    if i in d:
        d[i] += 1
    else:
        d[i] = 1
print(d)
```
