<h2>Decorator</h2>

```python
def outer(func):
    def inner(wrapper):
        return func(wrapper) * 2
    return inner

@outer
def wish(wrapper):
    return wrapper

obj = wish(7)
print(obj)
```

<h2>Generator</h2>

```python
def demo():
    yield "Hello, Good Morning !!"

obj = demo()
print(next(obj))
```
