<h3>random() function</h3>

```python
from random import *
```

```python
for i in range(10):
    print(random())
```

<h3>randint() function</h3>

```python
print(randint(1,5))
```

<h3>uniform() function</h3>

```python
print(uniform(1,3))
```

<h3>randrange([start],stop,[step])</h3>

```python
print(randrange(1,10,3))
```

```python
for i in range(10):
    print(randrange(1,11))
```

<h3>choice() function</h3>

```python
list = ['Sunny','Honey','Vinny']
for i in range(10):
    print(choice(list))
```
